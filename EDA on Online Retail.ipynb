## Onine Retail Transactions

Description: This dataset contains information about online retail transactions from an e-commerce company. It includes various attributes related to customer purchases, such as customer demographics, product details, transaction dates, and sales information.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

import warnings 
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Online Retail.csv')
df
```

```python
# Set 'InvoiceNo' as the index
df.set_index("InvoiceNo", inplace=True)
df.head(5)
```

```python
# Manipulate 'InvoiceDate' to extract only the year
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate']).dt.year
```

```python
df.head()
```

## Exploratory Data Analysis 

#### Analysis and Insights

```python
# Total quantity sold:

total_quantity_sold = df['Quantity'].sum()
print("Total Quantity Sold:", total_quantity_sold)
```

```python
# Maximum unit price:

max_unit_price = df['UnitPrice'].max()
print("Maximum Unit Price:", max_unit_price)
```

```python
# Average unit price:

avg_unit_price = df['UnitPrice'].mean()
print("Average Unit Price:", avg_unit_price)
```

```python
# Number of unique customers:

num_unique_customers = df['CustomerID'].nunique()
print("Number of Unique Customers:", num_unique_customers)
```

```python
# Total sales revenue:

df['TotalSales'] = df['Quantity'] * df['UnitPrice']
total_sales_revenue = df['TotalSales'].sum()
print("Total Sales Revenue:", total_sales_revenue)
```

```python
# Top 10 countries by sales revenue:

top_countries = df.groupby('Country')['TotalSales'].sum().nlargest(10)
print("Top 10 Countries by Sales Revenue:\n", top_countries)
```

```python
# Most frequently purchased products:

top_products = df['Description'].value_counts().nlargest(10)
print("Top 10 Most Frequently Purchased Products:\n", top_products)
```

#### Visualization 

```python
import matplotlib.pyplot as plt

plt.hist(df['Quantity'], bins=10, edgecolor='black')
plt.xlabel('Quantity')
plt.ylabel('Count')
plt.title('Distribution of Quantities Sold')
plt.show()
```

```python
revenue_by_country = df.groupby('Country')['TotalSales'].sum()

plt.figure(figsize=(12, 6))
revenue_by_country.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Sales Revenue')
plt.title('Sales Revenue by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_sales = df.groupby('InvoiceDate')['TotalSales'].sum()

plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o')
plt.xlabel('Month')
plt.ylabel('Sales Revenue')
plt.title('Monthly Sales Trend')
plt.xticks(rotation=45)
plt.show()
```

```python
top_customers = df.groupby('CustomerID')['TotalSales'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
top_customers.plot(kind='bar')
plt.xlabel('Customer ID')
plt.ylabel('Total Spending')
plt.title('Top 10 Customers by Total Spending')
plt.xticks(rotation=0)
plt.show()
```

```python
revenue_by_year = df.groupby('InvoiceDate')['TotalSales'].sum()

plt.figure(figsize=(10, 6))
revenue_by_year.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Sales Revenue')
plt.title('Sales Revenue by Year')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_price_year = df.groupby('InvoiceDate')['UnitPrice'].mean()

plt.figure(figsize=(10, 6))
avg_price_year.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Average Unit Price')
plt.title('Average Unit Price by Year')
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_sales_dist = df.groupby('InvoiceDate')['TotalSales'].sum()

plt.figure(figsize=(10, 6))
monthly_sales_dist.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Sales Revenue')
plt.title('Monthly Sales Distribution')
plt.xticks(rotation=0)
plt.show()
```

```python
top_products = df.groupby('Description')['Quantity'].sum().nlargest(10)

plt.figure(figsize=(10, 6))
top_products.plot(kind='bar')
plt.xlabel('Product Description')
plt.ylabel('Quantity Sold')
plt.title('Quantity Sold by Product Description (Top 10)')
plt.xticks(rotation=45)
plt.show()
```
