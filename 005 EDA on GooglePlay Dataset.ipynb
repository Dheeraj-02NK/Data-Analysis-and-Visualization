# EDA on Google Dataset

### Step 1: Import Libraries and Data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("googleplaystore.csv")
```

### Step 3: Data Manipulation

```python
# Convert "Price" column to numeric (remove '$' sign)
df['Price'] = df['Price'].str.replace('$', '')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
df.tail()
```

```python
# Map Type to numeric values
type_map = {
    'Free': 0,
    'Paid': 1
}
df['Type'] = df['Type'].map(type_map)
```

```python
df.head()
```

```python
df.head()
```

### Step 4: Visualization of Data

```python
# Count of apps in each category
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=df, order=df['Category'].value_counts().index)
plt.title('Number of Apps in Each Category')
plt.ylabel('Number of Apps')
plt.xlabel('Category')
plt.xticks(rotation=-90)
plt.show()
```

```python
# Distribution of app ratings
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='Rating', bins=20, kde=True)
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Content Rating distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Content Rating')
plt.title('Content Rating Distribution')
plt.xlabel('Content Rating')
plt.ylabel('Number of Apps')
plt.show()
```

```python
# Price distribution of apps
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Price', bins=20, kde=True)
plt.title('Price Distribution of Apps')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pairwise correlation heatmap of numeric columns
numeric_columns = ['Rating', 'Installs', 'Price', 'Content Rating', 'Type']
correlation_matrix = df[numeric_columns].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Pairwise Correlation Heatmap of Numeric Columns')
plt.show()
```

```python
# Price distribution by Content Rating
plt.figure(figsize=(10, 6))
sns.histplot(x='Content Rating', y='Price', data=df)
plt.title('Price Distribution by Content Rating')
plt.xlabel('Content Rating')
plt.ylabel('Price')
plt.show()
```

```python
# Distribution of app prices in the "Family" category
family_apps = df[df['Category'] == 'FAMILY']
plt.figure(figsize=(10, 6))
sns.histplot(x='Price', data=family_apps)
plt.title('Distribution of App Prices in the "Family" Category')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

### Step 5: Extracting Insights

```python
# Count of apps in each category
plt.figure(figsize=(10, 6))
sns.countplot(y='Category', data=df, order=df['Category'].value_counts().index)
plt.title('Number of Apps in Each Category')
plt.xlabel('Number of Apps')
plt.ylabel('Category')
plt.xticks(rotation=90)
plt.show()
```

```python
# Distribution of app installations
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Installs', bins=20, kde=True)
plt.title('Distribution of App Installations')
plt.xticks(rotation=45)
plt.xlabel('Installations')
plt.ylabel('Frequency')
plt.show()
```

```python
# Ratio of paid to free apps
plt.figure(figsize=(6, 6))
df['Type'].value_counts().plot.pie(autopct='%1.1f%%', labels=['Free', 'Paid'])
plt.title('Ratio of Paid to Free Apps')
plt.show()
```

```python
# Average Rating by Category
avg_rating_by_category = df.groupby('Category')['Rating'].mean()
print("Average Rating by Category:\n", avg_rating_by_category)
```

```python
# Paid vs. Free Apps
paid_vs_free_counts = df['Type'].value_counts()
print("Ratio of Paid to Free Apps:", paid_vs_free_counts[1] / paid_vs_free_counts[0])
```

```python
# Price and Rating Relationship
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Price', y='Rating')
plt.title('Price vs. Rating')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.show()
```

```python
# Content Rating vs. Rating
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Content Rating', y='Rating')
plt.title('Content Rating vs. Rating')
plt.xlabel('Content Rating')
plt.ylabel('Rating')
plt.show()
```

```python
# App Size and Installs
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Size', y='Installs')
plt.title('App Size vs. Installs')
plt.xlabel('Size')
plt.ylabel('Installs')
plt.show()
```

```python
# App Types and Ratings
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Type', y='Rating')
plt.title('App Type vs. Rating')
plt.xlabel('Type')
plt.ylabel('Rating')
plt.show()
```

```python
# Genres Analysis
top_genres = df['Genres'].value_counts().nlargest(5)
print("Top 5 Most Popular Genres:\n", top_genres)
```
