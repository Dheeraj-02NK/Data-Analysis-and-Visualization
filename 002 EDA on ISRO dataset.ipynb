## Data Analysis on ISRO Data

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

import warnings 
warnings.filterwarnings("ignore")
```

```python
# Load the dataset
df = pd.read_csv('ISRO Satellite Dataset.csv')
```

#### Visualization of Data

```python
satellite_count_by_operator = df['Operator/Owner'].value_counts()

# Plotting a bar chart
plt.figure(figsize=(10, 6))
satellite_count_by_operator.plot(kind='bar')
plt.title('Number of Satellites by Country/Operator')
plt.xlabel('Country/Operator')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
purpose_count = df['Purpose'].value_counts()

# Plotting a pie chart
plt.figure(figsize=(8, 8))
plt.pie(purpose_count, labels=purpose_count.index, autopct='%1.1f%%')
plt.title('Satellite Purpose Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Power (watts)'], bins=20, edgecolor='black')
plt.title('Power Distribution')
plt.xlabel('Power (watts)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Launch Mass (kg.)'], df['Power (watts)'])
plt.title('Satellite Launch Mass vs. Power')
plt.xlabel('Launch Mass (kg)')
plt.ylabel('Power (watts)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Expected Lifetime (yrs.)'], bins=10, edgecolor='black')
plt.title('Satellite Expected Lifetime Distribution')
plt.xlabel('Expected Lifetime (years)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Purpose'], df['Power (watts)'])
plt.title('Satellite Purpose vs. Power')
plt.xlabel('Purpose')
plt.ylabel('Power (watts)')
plt.xticks(rotation=45)
plt.show()
```

```python
orbit_count = df['Class of Orbit'].value_counts()

# Plotting a bar chart
plt.figure(figsize=(10, 6))
orbit_count.plot(kind='bar')
plt.title('Class of Orbit Distribution')
plt.xlabel('Class of Orbit')
plt.ylabel('Count')
plt.show()
```

```python
orbit_type_count = df['Type of Orbit'].value_counts()

# Plotting a bar chart
plt.figure(figsize=(10, 6))
orbit_type_count.plot(kind='bar')
plt.title('Type of Orbit Distribution')
plt.xlabel('Type of Orbit')
plt.ylabel('Count')
plt.show()
```

```python
df['Date of Launch'] = pd.to_datetime(df['Date of Launch'])
df['Launch Year'] = df['Date of Launch'].dt.year

plt.figure(figsize=(10, 6))
plt.hist(df['Launch Year'], bins=10, edgecolor='black')
plt.title('Satellite Launch Year Distribution')
plt.xlabel('Launch Year')
plt.ylabel('Count')
plt.show()
```

```python
launch_vehicle_count = df['Launch Vehicle'].value_counts()

# Plotting a bar chart
plt.figure(figsize=(10, 6))
launch_vehicle_count.plot(kind='bar')
plt.title('Satellite Launch Vehicle Distribution')
plt.xlabel('Launch Vehicle')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
launch_vehicle_power = df.groupby('Launch Vehicle')['Power (watts)'].mean()

plt.figure(figsize=(10, 6))
launch_vehicle_power.plot(kind='line', marker='o')
plt.title('Average Satellite Power by Launch Vehicle')
plt.xlabel('Launch Vehicle')
plt.ylabel('Average Power (watts)')
plt.xticks(rotation=45)
plt.show()
```

```python
contractor_count = df['Contractor'].value_counts()

# Plotting a bar chart
plt.figure(figsize=(10, 6))
contractor_count.plot(kind='bar')
plt.title('Satellite Contractor Distribution')
plt.xlabel('Contractor')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
launch_site_count = df['Launch Site'].value_counts()

# Plotting a bar chart
plt.figure(figsize=(10, 6))
launch_site_count.plot(kind='bar')
plt.title('Satellite Launch Site Distribution')
plt.xlabel('Launch Site')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 4))  # Adjust the figsize as per your preference
plt.scatter(df['Launch Site'], df['Power (watts)'])
plt.title('Satellite Launch Site vs. Power')
plt.xlabel('Launch Site')
plt.ylabel('Power (watts)')
plt.xticks(rotation=0)
plt.show()
```

```python
contractor_country_count = df['Country of Contractor'].value_counts()

# Plotting a bar chart
plt.figure(figsize=(10, 6))
contractor_country_count.plot(kind='bar')
plt.title('Satellite Country of Contractor Distribution')
plt.xlabel('Country of Contractor')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
